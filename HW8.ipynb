```python
data_source = 'phone_data.txt'
empty_str = 'Запись не выбрана'


def get_dat(dat, src):
    new_dat = []
    with open(src, 'r') as phdat:
        new_dat = phdat.read().split('\n')
    dat.clear()
    for entry in new_dat:
        dat.append(entry)

local_data = []
get_dat(local_data, data_source)
```

```python
# Все функции из этого блока берут на вход список строк справочника dat и текущую выбранную строку
# Все функции возвращают выбранную строку по итогам операции или empty_str

def spr_show(dat, sel_str):
    print()
    print('\n'.join(dat))
    return sel_str

def spr_add(dat, sel_str):
    nam = input('Введите имя контакта')
    num = input('Введите номер контакта')
    des = input('Введите описание контакта')
    new_str = ' '.join([nam,num,des])
    dat.append(new_str)
    return new_str

def spr_find(dat, sel_str):
    to_find = input('Введите данные для поиска')
    for record in dat:
        if to_find in record:
            return record
    return empty_str

def spr_remove(dat, sel_str):
    dat.remove(sel_str)
    return empty_str

def spr_change(dat, sel_str):
    spr_remove(dat, sel_str)
    return spr_add(dat, sel_str)

def spr_save(dat, sel_str):
    data_source = input('Введите название справочника назначения')
    with open(data_source, 'w') as phdat:
        phdat.write('\n'.join(dat))
    return sel_str

def spr_paste(dat, sel_str):
    dat.append(sel_str)
    return sel_str

def spr_sprchng(dat, sel_spr):
    data_source = input('Введите название справочника назначения')
    get_dat(dat, data_source)
    return sel_spr

def spr_exit(dat, sel_str):
    return empty_str

func_dict = {1: spr_show,
             2: spr_add,
             3: spr_find,
             4: spr_change,
             5: spr_remove,
             6: spr_save,
             8: spr_paste,
             9: spr_sprchng,
             0: spr_exit}
```

```python
# Main programm


actions_desc = [
    '1. Показать все контакты',
    '2. Создать контакт',
    '3. Найти контакт',
    '4. Изменить контакт',
    '5. Удалить контакт',
    '6. Сохранить изменения',
    '8. Вставить текущую строку',
    '9. Выбрать другой справочник',
    '0. Выйти из справочника'
]

def input_request(curr_contact):
    print(f'\n Выбранная строка: \n { curr_contact} \n')
    for act in actions_desc:
        print(act)
    return(int(input('Выберите цифровой код действия')))

sel_data = empty_str
todo = 1
while todo:
    todo = input_request(sel_data)
    sel_data = func_dict[todo](local_data, sel_data)
```
